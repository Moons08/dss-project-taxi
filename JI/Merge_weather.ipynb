```python
weather1.head()
```

```python
weather1["pickup_datetime"] = pd.to_datetime(weather1["pickup_datetime"]) # 날짜 데이터로 변환
```

```python
# 51분 간격대의 시간대만 가져오기 위함
ls1 = []
for num in range(0, 10481):
    if weather1["pickup_datetime"][num].minute == 51:
        ls1.append(weather1["pickup_datetime"][num])
    else:
        pass
```

```python
ls2 = []
for num in range(0, 10481):
    if weather1["pickup_datetime"][num].minute == 51:
        ls2.append(weather1["tempm"][num])
    else:
        pass
```

```python
ls3 = []
for num in range(0, 10481):
    if weather1["pickup_datetime"][num].minute == 51:
        ls3.append(weather1["rain"][num])
    else:
        pass    
```

```python
ls4 = []
for num in range(0, 10481):
    if weather1["pickup_datetime"][num].minute == 51:
        ls4.append(weather1["snow"][num])
    else:
        pass 
```

```python
weather_data  = pd.DataFrame(columns = ["pickup_datetime", "tempm", "rain", "snow"])
weather_data["pickup_datetime"] = ls1
weather_data["tempm"] = ls2
weather_data["rain"] = ls3
weather_data["snow"] = ls4
```

```python
weather_date = weather_data[25:4223] # 우리 데이터에 있는 날짜만 가져옴
weather_date.reset_index(drop=True).tail()
```
